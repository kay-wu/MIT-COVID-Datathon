```
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

### Load ZCTA list

```
## List of cities
cities = ['NY_covid', 
          'alexandria_covid', 
          'chicago_covid', 
          'houston_covid', 
          'st_louis_covid']
```

### Scraping Demographic Info

```
for city in cities:
  # get zcta list and number of covid cases per zcta
  city_covid = pd.read_csv('https://raw.githubusercontent.com/kay-wu/MIT-COVID-Datathon/master/data/covid_data/processed/' + city + '.csv')
  zcta = city_covid['zcta']
  covid_cases = city_covid['covid_cases']
  print('[', city, '] zcta length', len(zcta))


  # start scraping
  demographic_df = pd.DataFrame()
  demographic_percent_df = pd.DataFrame()
  for idx, zip in enumerate(zcta):
    url = 'http://mcdc.missouri.edu/applications/acs/profiles/report.php?period=5&year=2018&g=86000US' + str(int(zip))
    #print(idx, url, covid_cases[idx])
    html_content = requests.get(url).text

    # Parse the html content
    soup = BeautifulSoup(html_content, "lxml")
    column_names = ['zipcode','covid_cases']
    number_values = [int(zip),covid_cases[idx]]
    percent_values = [int(zip),covid_cases[idx]]
    table_demographic = soup.find('table', id="grptbl-Demographic")
    subjects = table_demographic.find_all("tr", class_="subjectHeadRow")
    subjects[0].find('th').find(text=True, recursive=False)

    rows = table_demographic.find_all("tr")
    prefix = ''
    for index, row in enumerate(rows):
        if row.has_attr('class') and row['class'][0] == 'headRow':
          pass
        elif row.has_attr('class') and row['class'][0] == 'subjectHeadRow':
          subject = row.find('th').find(text=True, recursive=False)
          prefix = subject.strip() + '.'
        else:
          tds = row.find_all('td')
          if(len(tds) > 1):
            field0 = prefix + tds[0].text 
            field1 = tds[1].text
            field2 = None
            if len(tds) > 2:
              field2 = tds[2].text
            column_names.append(field0)
            number_values.append(field1)
            percent_values.append(field2)
          else:
            #print('error: ' + str(zip))
            pass
    
    if( len(column_names) > 2 ):
      demo_df = pd.DataFrame([number_values], columns=column_names)
      demographic_df = demographic_df.append(demo_df)

      demo_percent_df = pd.DataFrame([percent_values], columns=column_names)
      demographic_percent_df = demographic_percent_df.append(demo_percent_df)

  demographic_df.to_csv('demographic_number_info_' + city + '.csv')
  demographic_percent_df.to_csv('demographic_percent_info_' + city + '.csv')
  print('===> ', city, 'finish demographic total info: ', len(demographic_df))
```
